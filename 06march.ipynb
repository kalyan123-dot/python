```python
x = df[['width','height']]
y = df['fruit_label']
knnCls2 = KNeighborsClassifier()
knnCls2.fit(x,y)
knnCls2.score(x,y)
```

```python
neighbours = [1,2,3,4,8,32]
scores = {}

for n in neighbours:
    cls = KNeighborsClassifier(n_neighbors=n)
    cls.fit(X,y)
    accuracy = cls.score(X,y)
    scores[n] = accuracy
print(scores)
```
